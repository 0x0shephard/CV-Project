# Week 3: Multi-View SfM & Bundle Adjustment

This notebook demonstrates the complete incremental Structure from Motion pipeline:
1. Initialize reconstruction from two views
2. Incrementally add new views using PnP (Perspective-n-Point)
3. Observe reconstruction drift
4. Apply Bundle Adjustment for global refinement
5. Compare before/after results

**Date:** December 2025

```python
# Import necessary libraries
import sys
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Add src directory to path
sys.path.append(os.path.abspath('../src'))

# Import our SfM modules
from sfm_pipeline import SfMPipeline

print("All imports successful!")
print(f"OpenCV version: {cv2.__version__}")
print(f"NumPy version: {np.__version__}")
```

## 1. Setup and Configuration

Configure the pipeline parameters and paths.

```python
# Configuration
IMAGE_DIR = "../Data"  # Directory containing your images
OUTPUT_DIR = "../output/week3"  # Output directory for results
IMAGE_PATTERN = "*.jpg"  # Pattern for image files
MAX_IMAGES = 10  # Process first 10 images (adjust as needed)
DETECTOR_TYPE = "SIFT"  # Feature detector: 'SIFT' or 'ORB'

# Create output directory
os.makedirs(OUTPUT_DIR, exist_ok=True)

print(f"Configuration:")
print(f"  Image directory: {IMAGE_DIR}")
print(f"  Output directory: {OUTPUT_DIR}")
print(f"  Max images: {MAX_IMAGES}")
print(f"  Detector: {DETECTOR_TYPE}")
```

## 2. Initialize SfM Pipeline

Create the pipeline and load images.

```python
# Create pipeline
pipeline = SfMPipeline(IMAGE_DIR, OUTPUT_DIR)

# Load images
pipeline.load_images(pattern=IMAGE_PATTERN, max_images=MAX_IMAGES)

print(f"\nLoaded {len(pipeline.images)} images")
print(f"Image dimensions: {pipeline.images[0].shape}")
print(f"\nCamera Intrinsic Matrix K:")
print(pipeline.K)
```

## 3. Two-View Initialization

Start by reconstructing from the first two images.

```python
# Initialize from first two images
success = pipeline.initialize_two_view(image_id1=0, image_id2=1, detector_type=DETECTOR_TYPE)

if success:
    print("\n✓ Two-view initialization successful!")
    
    # Get initial point cloud
    points_init, colors_init = pipeline.sfm.get_point_cloud()
    
    print(f"\nInitial reconstruction:")
    print(f"  Cameras: {len(pipeline.sfm.camera_poses)}")
    print(f"  3D Points: {len(points_init)}")
else:
    print("\n✗ Two-view initialization failed!")
```

### Visualize Initial Two-View Reconstruction

```python
from visualization import visualize_reconstruction

# Get camera poses
camera_poses = [(cam.R, cam.t) for cam in pipeline.sfm.camera_poses.values()]

# Visualize
fig, ax = visualize_reconstruction(
    points_init, 
    colors_init, 
    camera_poses=camera_poses,
    title="Initial Two-View Reconstruction",
    point_size=2.0
)
plt.show()
```

## 4. Incremental View Addition with PnP

Now add the remaining images one by one using Perspective-n-Point (PnP) algorithm.

### How PnP Works:
1. Match 2D features in new image with existing 3D map points
2. Use RANSAC-based PnP to estimate camera pose
3. Triangulate new 3D points between new view and previous views
4. Add new points to the map

```python
# Add remaining views incrementally
if len(pipeline.images) > 2:
    pipeline.add_views_incrementally(start_idx=2)
    
    print("\n" + "="*60)
    print("RECONSTRUCTION AFTER INCREMENTAL SfM")
    print("="*60)
    print(f"Total cameras: {len(pipeline.sfm.camera_poses)}")
    print(f"Total 3D points: {len(pipeline.sfm.map_points)}")
    
    # Get point cloud before BA
    points_before_ba, colors_before_ba = pipeline.sfm.get_point_cloud()
else:
    print("Only 2 images available, skipping incremental addition")
```

### Visualize Reconstruction Before Bundle Adjustment

Notice potential drift and inconsistencies due to accumulated errors.

```python
from visualization import visualize_multiple_views, plot_camera_trajectory

if len(pipeline.images) > 2:
    # Get camera poses
    camera_poses_before = [(cam.R, cam.t) for cam in pipeline.sfm.camera_poses.values()]
    
    # Multiple views
    fig1 = visualize_multiple_views(
        points_before_ba,
        colors_before_ba,
        camera_poses=camera_poses_before,
        title="Reconstruction BEFORE Bundle Adjustment"
    )
    plt.show()
    
    # Camera trajectory
    fig2 = plot_camera_trajectory(
        camera_poses_before,
        title="Camera Trajectory (Before BA)"
    )
    plt.show()
```

## 5. Bundle Adjustment Refinement

Apply Bundle Adjustment to globally refine all camera poses and 3D points.

### What is Bundle Adjustment?
Bundle Adjustment (BA) is a non-linear least squares optimization that:
- Simultaneously adjusts all camera poses and 3D point positions
- Minimizes reprojection error across all observations
- Corrects accumulated drift from incremental reconstruction

This is the gold standard for refining Structure from Motion reconstructions.

```python
if len(pipeline.images) > 2:
    # Run Bundle Adjustment
    ba_info = pipeline.run_bundle_adjustment(fix_first_camera=True)
    
    print("\n" + "="*60)
    print("BUNDLE ADJUSTMENT RESULTS")
    print("="*60)
    print(f"Initial RMSE: {ba_info['initial_rmse']:.4f} pixels")
    print(f"Final RMSE: {ba_info['final_rmse']:.4f} pixels")
    print(f"Improvement: {ba_info['improvement']:.4f} pixels")
    print(f"Success: {ba_info['success']}")
    print(f"Iterations: {ba_info['iterations']}")
    
    # Get refined point cloud
    points_after_ba, colors_after_ba = pipeline.sfm.get_point_cloud()
```

### Visualize Reconstruction After Bundle Adjustment

```python
if len(pipeline.images) > 2:
    # Get refined camera poses
    camera_poses_after = [(cam.R, cam.t) for cam in pipeline.sfm.camera_poses.values()]
    
    # Multiple views
    fig3 = visualize_multiple_views(
        points_after_ba,
        colors_after_ba,
        camera_poses=camera_poses_after,
        title="Reconstruction AFTER Bundle Adjustment"
    )
    plt.show()
    
    # Camera trajectory
    fig4 = plot_camera_trajectory(
        camera_poses_after,
        title="Camera Trajectory (After BA)"
    )
    plt.show()
```

## 6. Before/After Comparison

Side-by-side comparison of the reconstruction quality.

```python
from visualization import compare_reconstructions

if len(pipeline.images) > 2:
    fig5 = compare_reconstructions(
        points_before_ba,
        colors_before_ba,
        points_after_ba,
        colors_after_ba
    )
    plt.show()
    
    print("\n" + "="*60)
    print("COMPARISON")
    print("="*60)
    print(f"Points before BA: {len(points_before_ba)}")
    print(f"Points after BA: {len(points_after_ba)}")
    print(f"Reprojection error improved by: {ba_info['improvement']:.4f} pixels")
```

## 7. Save Results

Export the final refined reconstruction to PLY format and save visualizations.

```python
# Save all results
if len(pipeline.images) > 2:
    pipeline.save_results(ba_info=ba_info)
    pipeline.visualize_results()
else:
    pipeline.save_results(ba_info=None)

print("\n" + "="*60)
print("ALL RESULTS SAVED")
print("="*60)
print(f"Output directory: {OUTPUT_DIR}")
print("\nGenerated files:")
print("  - reconstruction_two_view.ply (initial 2-view reconstruction)")
if len(pipeline.images) > 2:
    print("  - reconstruction_before_ba.ply (before bundle adjustment)")
    print("  - reconstruction_after_ba.ply (after bundle adjustment)")
    print("  - ba_comparison.png (before/after comparison)")
print("  - reconstruction_refined.ply (final result)")
print("  - reconstruction_multiview.png (multiple viewpoints)")
print("  - camera_trajectory.png (camera path)")
print("  - reconstruction_summary.txt (statistics)")
```

2. **Incremental SfM with PnP**: 
   - Each new image localized using 2D-3D correspondences
   - New 3D points triangulated and added to map
   - Accumulated errors caused drift over time

3. **Bundle Adjustment**:
